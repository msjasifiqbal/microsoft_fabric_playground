```python
%%sql
--Find out which city hosted most of matches, then sort and get the top 5 cities
SELECT info_city, COUNT(info_city) AS MATCHCNT FROM ipl_summary_raw 
GROUP BY info_city
ORDER BY MATCHCNT DESC
LIMIT 5
```

```python
%%sql
--Find out which team won the matches and count the number of times
SELECT info_outcome_winner AS City, COUNT(info_outcome_winner) AS WinCount FROM ipl_summary_raw
GROUP BY info_outcome_winner
ORDER BY WinCount DESC
```

```python
%%sql
--Find out which team did make to playoff stage and how many times
SELECT COL1 AS Teams,COUNT(COL1) AS PlayoffCount FROM 
  (
    SELECT info_teams_1 AS COL1 FROM ipl_summary_raw 
    WHERE info_event_stage IN ('Qualifier 1','Qualifier 2','Eliminator')
    UNION ALL
    SELECT info_teams_2 AS COL2 FROM ipl_summary_raw
    WHERE info_event_stage IN ('Qualifier 1','Qualifier 2','Eliminator')
  ) TBL1
    GROUP BY COL1
    ORDER BY PlayoffCount DESC 


    
--WHERE info_event_stage IN ('Qualifier 1','Qualifier 1','Semi Final')

```

```python
%%sql
--Find out the match officials for Final matches

SELECT DISTINCT(info_officials_match_referees_1),Designation AS MatchOfficials FROM 
(
SELECT info_officials_match_referees_1, 'info_officials_match_referees_1' AS Designation FROM ipl_summary_raw WHERE info_event_stage = 'Final'
UNION ALL
SELECT	info_officials_reserve_umpires_1,'info_officials_reserve_umpires_1' AS Designation FROM ipl_summary_raw WHERE info_event_stage = 'Final'
UNION ALL
SELECT info_officials_tv_umpires_1,'info_officials_tv_umpires_1' AS Designation	FROM ipl_summary_raw  WHERE info_event_stage = 'Final'
UNION ALL
SELECT info_officials_umpires_1,'info_officials_umpires_1' AS Designation	FROM ipl_summary_raw  WHERE info_event_stage = 'Final'
UNION ALL
SELECT info_officials_umpires_2,'info_officials_umpires_2' Designation  FROM ipl_summary_raw  WHERE info_event_stage = 'Final'
) TBL1
```

```python
%%sql
--Find out which venue hosted final match with year and city name
SELECT info_venue,info_city,info_season FROM ipl_summary_raw WHERE info_event_stage = 'Final'
```

```python
%%sql
--Find out which are the match got tie.
SELECT info_teams_1,info_teams_2 FROM ipl_summary_raw WHERE info_outcome_result = 'tie'
```

```python
%%sql
--Find out which team made to final did not win the match

SELECT DISTINCT T1 FROM(
SELECT info_teams_1 AS T1 from ipl_summary_raw where info_event_stage = 'Final' 
UNION 
SELECT info_teams_2 AS T1 from ipl_summary_raw where info_event_stage = 'Final'
    ) TBL1 WHERE T1 NOT IN(SELECT DISTINCT info_outcome_winner FROM ipl_summary_raw where info_event_stage = 'Final')
```
